```python
# train and test split 
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size = 0.2, random_state = 1)

# random_state =1 ......so every time it will generate same train and test set
```

```python
X_train_data_frame= train_data.iloc[:,0:4]
X_train_data_frame
```

```python
X_train = train_data.iloc[:,0:4].values
Y_train = train_data.iloc[:,4].values


X_test = test_data.iloc[:,0:4].values
Y_test = test_data.iloc[:,4].values
```

```python
len(Y_test)
```

# Apply Linear SVM (Support Vector Machine) 

```python
from sklearn.svm import SVC
classifier=SVC(kernel="linear")
classifier.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
Y_pred = classifier.predict(X_test)
accuracy_score(Y_test, Y_pred)
```

here linear svm classify those datas into different classes with accuracy 100% 

# SVM  (Support Vector Machine)

  here c=100 ... which is a very high value for C....So it will suffer from overfitting problem

```python
from sklearn.svm import SVC
classifier = SVC(C=100,random_state=1)  # linear svm applied
classifier.fit(X_train,Y_train)
```

```python
print(classifier.support_vectors_)

Y_pred = classifier.predict(X_test)

```

```python
test_data["Predictions"] = Y_pred
```

```python
test_data
```

in this above test data  77th data row is wrongly predicted., as its a overfitting problem

```python
from sklearn.metrics import accuracy_score
Y_pred = classifier.predict(X_test)
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
print(cm)
print(classification_report(Y_test, Y_pred))
accuracy = float(cm.diagonal().sum())/len(Y_test)     #   cm.diagonal()=(11+12+6)  and   len(Y_test)=30
print("\nAccuracy Of SVM For IRIS Dataset : ", accuracy)
```

the accuracy is not 1 as we choose high C value ...becuse of that it suffers from overfitting problem
