```python
# here i am choosing 3rd column for x values and 4th column for y values 
X = d.iloc[:, 2]
Y = d.iloc[:, 4]
#X = d['GrLivArea']
#Y = d['SalePrice']
X=(X - X.mean()) / X.std()
#Y=(Y - Y.mean()) / Y.std()
plt.scatter(X, Y)
# no need to write down plt.show() as we already specify %matplotlib inline 
#plt.show() 
print(X.size)
```

```python
# Building the model
# we can choose any random value for theta0 and theta1
# y_predict=theta0+theta1*x
theta1 = 0
theta0 = 0
# we can choose nay random value for learning rate
L_R = 0.01  # The learning Rate(minimum theoritical value for Learning rate is 0)
iters = 1000 # The number of iterations to perform gradient descent(its the epoch value)

m = X.size # Number of elements in X (number of rows in the x feature colume ie, no of elements i)
J=np.zeros(iters) #to store cost  # used for storing the cost in the array  with all elements 0
k=np.zeros(iters) # to store the iteration number # used for storing the number of iterations in the array  with all elements 0
parameters_0 = np.zeros(iters) # used for storing the theta0 value in the array  with all elements 0
parameters_1 = np.zeros(iters) # used for storing the theta1 in the array  with all elements 0
# Gradient Descent Algorithm
for i in range(iters): 
    Y_pred = theta1*X + theta0  # The current predicted value of Y
    J[i]=(1/2*m)*sum(np.square(Y_pred-Y))  # cost function at a particular iteration
    Grad_theta1 = (2/m) * sum(X * (Y_pred-Y))  # Partial Derivative wrt theta1
    Grad_theta0 = (2/m) * sum(Y_pred-Y)  # Partial Derivative wrt theta0
    theta1=theta1- L_R * Grad_theta1  # Update m
    theta0 = theta0 - L_R * Grad_theta0  # Update c
    k[i]=i   # insert into the iteration array with the present iteration number
    parameters_0[i]=theta0  # insert into the parameters_0 array with the present theta0 number
    parameters_1[i]=theta1  # insert into the parameters_1 array with the present theta1 number
    print("theta0= {} , theta1= {} , cost= {} and epoch= {} ".format(theta0,theta1,J[i],i))
    
#print (theta1, theta0)
#print(parameters_0) 
```

```python
min(parameters_1)
```

```python
min(parameters_0)
```

```python
min(J)
```

```python
# Making predictions
Y_pred = theta1*X + theta0

plt.scatter(X, Y)
#plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.plot(X, Y_pred, color='red') # predicted

#plt.show()
```

```python
# plot between epoch (ie, the number of iteration vs cost )
plt.plot(k, J, color='red')
plt.xlabel('K(EPOCH)')
plt.ylabel('J(COST)')
plt.title('Change in COST(J) with change in EPOCH(k)')
plt.grid(True)
```

```python
plt.scatter(parameters_0, J, color='blue')
plt.xlabel('parameters_0')
plt.ylabel('J(COST)')
plt.title('Change in theta0 with change in J(COST)')
plt.grid(True)
```

```python
plt.scatter(parameters_1, J, color='orange')
plt.xlabel('parameters_1')
plt.ylabel('J(COST)')
plt.title('Change in theta1 with change in J(COST)')
plt.grid(True)
```
