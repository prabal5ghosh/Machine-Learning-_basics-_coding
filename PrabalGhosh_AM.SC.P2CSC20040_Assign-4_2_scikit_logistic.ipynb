# USE THE data_classification FILE FOR LOGISTIC REGRESSION

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
exams = pd.read_csv("D:\\semester 3 videos\\Machine Learning\\notes\\Lab\\Lab4_06_09_Logistic\\data_classification.csv",names=['Studied','Slept','Passed'])
fig = plt.figure()
ax = fig.add_subplot(111)
colors = ['red', 'blue']
ax.scatter(exams.Studied, exams.Slept, s=25, marker="o",  c=exams['Passed'], cmap=matplotlib.colors.ListedColormap(colors))
```

```python
sns.scatterplot(data=exams, x="Studied", y="Slept",hue="Passed")
```

```python
sns.scatterplot(data=exams, x="Studied", y="Slept",hue="Passed",style="Passed")
```

```python
from sklearn.linear_model import LogisticRegression
features = exams.drop(['Passed'],axis=1)
target = exams['Passed']
logmodel = LogisticRegression()
logmodel.fit(features, target)
predictions = logmodel.predict(features)
```

```python
print('intercept:',logmodel.intercept_[0])
print('coefficients:',logmodel.coef_)
```

```python
exams['boundary'] = (-logmodel.intercept_[0] - (logmodel.coef_[0][0] * features['Studied'])) / logmodel.coef_[0][1]
```

```python
plt.scatter(exams['Studied'],exams['Slept'], s=25, marker="o", c=exams['Passed'], cmap=matplotlib.colors.ListedColormap(colors))
plt.plot(exams['Studied'], exams['boundary'])
plt.show()
```

```python
sns.scatterplot(data=exams, x="Studied", y="Slept",hue="Passed",style="Passed")
plt.plot(exams['Studied'], exams['boundary'])
```

# USE THE Social_Network_Ads.csv FILE FOR LOGISTIC REGRESSION

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
exams = pd.read_csv("D:\\semester 3 videos\\Machine Learning\\notes\\Lab\\Lab4_06_09_Logistic\\Social_Network_Ads.csv")
fig = plt.figure()
ax = fig.add_subplot(111)
colors = ['red', 'blue']
ax.scatter(exams.Age, exams.EstimatedSalary, s=25, marker="o",  c=exams['Purchased'], cmap=matplotlib.colors.ListedColormap(colors))
```

```python
exams.head()
```

```python
sns.scatterplot(data=exams, x="Age", y="EstimatedSalary",hue="Purchased",style="Purchased")
```

```python
from sklearn.linear_model import LogisticRegression
features = exams.drop(['Purchased','Gender','User ID'],axis=1)
target = exams['Purchased']
logmodel = LogisticRegression()
logmodel.fit(features, target)
predictions = logmodel.predict(features)
```

```python
print('intercept:',logmodel.intercept_[0])
print('coefficients:',logmodel.coef_)
```

```python
exams['boundary'] = (-logmodel.intercept_[0] - (logmodel.coef_[0][0] * features['Age'])) / logmodel.coef_[0][1]
```

```python
plt.scatter(exams['Age'],exams['EstimatedSalary'],s=25, marker="o", c=exams['Purchased'], cmap=matplotlib.colors.ListedColormap(colors))
plt.plot(exams['Age'], exams['boundary'])
plt.show()
```
