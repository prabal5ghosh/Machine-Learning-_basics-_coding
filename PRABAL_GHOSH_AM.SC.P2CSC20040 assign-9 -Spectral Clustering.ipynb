```python
from sklearn.cluster import SpectralClustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

iris=pd.read_csv("C:/Users/Prabal Ghosh/Desktop/iris.csv")

iris1=iris.iloc[:,0:4]
y=iris.iloc[:,-1]
plt.scatter(iris1.iloc[:,0], iris1.iloc[:,2])
plt.show()

sc = SpectralClustering(n_clusters=3).fit(iris1)
print(sc)
 
labels = sc.labels_
plt.scatter(iris1.iloc[:,0], iris1.iloc[:,2], c=labels)
plt.show()

f = plt.figure()
f.add_subplot(2, 2, 1)
for i in range(2, 5):
 sc = SpectralClustering(n_clusters=i).fit(iris1)
 f.add_subplot(2, 2, i-1)
 plt.scatter(iris1.iloc[:,0], iris1.iloc[:,2], s=5, c=sc.labels_, label="n_cluster-"+str(i))
 plt.legend()

plt.show() 
```
